# Histogram with Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style(style="darkgrid")
```

```python
df = pd.read_csv("Piece_Dim.csv", index_col="Item_No")
df.head()
```

```python
df.describe()
```

# Histogram

```python
plt.figure(figsize=(10, 10))
plt.hist(x=df["Length"])
plt.xlabel("Length in mm")
plt.ylabel("Count")
plt.title("Histogram of piece plot")
plt.show()
```

# Box Plot

```python
plt.figure(figsize=(10, 8))
plt.boxplot(x=df["Length"])
plt.show()
```

```python
op1 = df[df["Operator"] == "Op-1"]
op2 = df[df["Operator"] == "Op-2"]
```

```python
plt.figure(figsize=(10, 8))
plt.boxplot([op1["Length"], op2["Length"]], notch=True, )
plt.xlabel("Operator")
plt.ylabel("Length")
plt.title("Box Plot by Operator")
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.boxplot(data=[op1["Length"], op2["Length"]])
plt.xlabel("Operator")
plt.ylabel("Length")
plt.title("Box Plot by Operator")
plt.show()
```

# Lineplot

```python
sns.lineplot(x=np.arange(0, len(df)), y=df["Length"])
mean_l = df["Length"].mean()
std_l = df["Length"].std()
plt.axhline(mean_l, color="red", linestyle="-")
plt.axhline(mean_l + 3*std_l, color="blue", linestyle="-", label="Mean")
plt.axhline(mean_l - 3*std_l, color="blue", linestyle="-")
plt.legend()
plt.show()
```

# Scatterplot

```python
sns.scatterplot(x=df["Length"], y=df["Width"], marker="D")
plt.show()
```

# Bar Plot

```python
opers = ["Operator1", "Operator2", "Operator3"]
pieces = [50, 70, 45]
plt.bar(opers, pieces)
plt.show()
```

```python
df.head()
```

```python
low = df[df["Length"] <= 95]["Length"].count()
medium = df[(df["Length"] > 95) & (df["Length"] <= 105)]["Length"].count()
high = df[df["Length"] > 105]["Length"].count()
```

```python
sizes = [low, medium, high]
ranges = ["x <= 95", "95 < x <= 105", "x > 105"]
plt.bar(ranges, sizes)
plt.show()
```

# Saving plots as images

```python
sizes = [low, medium, high]
ranges = ["x <= 95", "95 < x <= 105", "x > 105"]
plt.bar(ranges, sizes)
plt.savefig("myfig.png")
plt.savefig("myfig.jpg")
```
