# Titanic Dataset Explore

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style(style="darkgrid")
import numpy as np
```

```python
sns.get_dataset_names()
```

```python
t = sns.load_dataset("titanic")
```

```python
t.head()
```

```python
t.describe()
```

```python
t.info()
```

```python
sns.displot(data=t, x="age", kde=True, col="sex", row="survived")
```

```python
sns.catplot(data=t, x="survived", y="age", kind="swarm", aspect=2)
```

```python
sns.catplot(data=t, x="survived", y="age", kind="box", aspect=2)
```

```python
sns.catplot(data=t, x="sex", kind="count", hue="survived")
```

```python
sns.catplot(data=t, x="deck", kind="count", hue="survived")
```
