# Has the variance of the population changed based on the sample?
## Fromally this is known as a Chi-squared test which tests the population variance against a specified value.


$$
H_0: s^2 = \sigma^2 \\
H_a: s^2 \neq \sigma^2
$$

The test statistic is

$$\chi^2 = \frac{(n-1)s^2}{\sigma^2}$$

Example: 51 bottles taken from our machine, we find std dev of 2.35cc, has it <b><u>increased</u></b> from the established 2cc level? (use 90% confidence, thus $\alpha = 0.1$) <br />
We also have degreed of freedom in this test, but python takes care of it given the sample

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
import scipy.stats as stats
from statsmodels.stats import weightstats
```

```python
# plot chi-squared

x_ax = np.linspace(0, 100, 101)
y_ax = stats.chi2.pdf(x_ax, df=20) # df -> degrees of freedom

sns.relplot(x=x_ax, y=y_ax, kind="line")
plt.plot()
```

```python
# no built in chi-squared test, have to do it manually

# calc test statistic
chi_sq = (51-1)*2.35**2/2**2
chi_sq
```

```python
# calculate the critical value
stats.chi2.isf(0.1, 50)
```

```python
# chi_sq > critical value hence we reject the null hypothesis (one-tail)
```
