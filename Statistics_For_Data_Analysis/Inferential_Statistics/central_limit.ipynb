```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
```

# Roll a Die 1000000 Times

```python
dice_rolls = np.random.randint(1, 7, 1000000)
```

```python
sns.displot(x=dice_rolls, bins=6)
plt.show()
```

```python
np.mean(dice_rolls)
```

```python
np.std(dice_rolls)
```

# Take a sample of the dice_rolls
## Take a big sample and bootstrap it

```python
sample = np.random.choice(dice_rolls, 100000)
```

```python
sample_means = []
for i in range(100000):
    # 10 in a bootstrap sample
    s = np.random.choice(sample, 10)
    sample_means.append(s.mean())
sample_means = np.array(sample_means)
```

```python
sns.displot(data=sample_means, bins=20)
plt.xlim((0, 6))
```

# No bootstrapping

```python
sample_means = []
for i in range(1000):
    s = np.random.choice(dice_rolls, 100)
    sample_means.append(s.mean())
sample_means = np.array(sample_means)
```

```python
sns.displot(data=sample_means)
plt.xlim((0, 6))
```

# The dice rolls constitute a uniform distribution, what about other distributions

```python
a = np.random.normal(100, 5, 50000)
b = np.random.normal(80, 5, 30000)
bm = np.append(a, b)
np.random.shuffle(bm)
bm.mean()
```

```python
sns.displot(bm, kind="kde")
plt.show()
```

```python
# clearly bimodal
sns.displot(bm)
plt.show()
```

```python
sample_means = []
for i in range(10000):
    s = np.random.choice(bm, 100)
    sample_means.append(s.mean())
sample_means = np.array(sample_means)
```

```python
sns.displot(data=sample_means)
plt.xlim((80, 100))
```
