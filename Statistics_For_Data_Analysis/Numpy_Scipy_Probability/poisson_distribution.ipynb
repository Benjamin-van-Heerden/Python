# Poisson NumPy

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="whitegrid")
```

```python
people_in_queue = np.random.poisson(3.6, 1000000)
print(sum(people_in_queue == 7)/len(people_in_queue))
```

```python
sns.countplot(x=people_in_queue)
plt.show()
```

# Poisson SciPy

```python
from scipy.stats import poisson
```

```python
k = 7
mu = 3.6
poisson.pmf(k, mu)
```

```python
poisson.cdf(7, 3.6)
```

```python
x_ax = np.arange(0, 15)
y_ax = poisson.pmf(x_ax, 3.6)
sns.barplot(x=x_ax, y=y_ax)
plt.show()
```

```python
x_ax = np.arange(0, 15)
y_ax = poisson.cdf(x_ax, 3.6)
sns.barplot(x=x_ax, y=y_ax)
plt.show()
```
