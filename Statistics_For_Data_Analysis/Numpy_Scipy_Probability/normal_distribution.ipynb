# Simulate using numpy

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="whitegrid")
```

```python
np.random.normal(100, 2)
```

```python
sample = np.random.normal(100, 2, 1000000)
sns.histplot(x=sample, kde=True)
plt.show()
```

```python
#within 1sigma, 2sigma, 3sigma
print(sum((sample > 98) & (sample < 102))/len(sample))
print(sum((sample > 96) & (sample < 104))/len(sample))
print(sum((sample > 94) & (sample < 106))/len(sample))
```

# Scipy for theoretical probabilities

```python
from scipy.stats import norm
```

## Bottles filled with 150cc with sigma 2cc. What percent of bottles have volume greater than 153cc

```python
norm.sf(153, 150, 2)
```

## What percent of bottles have volume between 148 and 152cc

```python
norm.cdf(152, 150, 2) - norm.cdf(148, 150, 2) 
```

```python
x_ax = np.linspace(142, 158, 10000)
y_ax = norm.pdf(x_ax, 150, 2)
sns.lineplot(x=x_ax, y=y_ax)
plt.axvline(148, color="red")
plt.axvline(152, color="red")
plt.show()
```
