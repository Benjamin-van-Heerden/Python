# A manufacturer has 12% defect rate in production. Buyer decides to test 20 random pieces and will accept if there are two or less defectives. What is the probability of getting accepted?

```python
two_rej = np.random.binomial(20, 0.12, 10000)
print(sum(two_rej <= 2 )/len(two_rej))
```

```python
sns.countplot(x=two_rej)
plt.show()
```

# SciPy for exact values

```python
from scipy.stats import binom
```

```python
#two or less
k = 2
n = 20
p = 0.12
binom.cdf(k, n, p)
```

```python
#more than two
print(1 - binom.cdf(2, 20, 0.12))
#or
print(binom.sf(2, 20, 0.12))
```

```python
#exactly two
binom.pmf(2, 20, 0.12)
```

```python
binom.mean(20, 0.12)
```

```python
binom.var(20, 0.12)
```

```python
binom.std(20, 0.12)
```

```python
x_axis = np.arange(0, 21)
y_axis = binom.pmf(x_axis, 20, 0.12)
sns.barplot(x=x_axis, y=y_axis)
plt.show()
```

```python
x_axis = np.arange(0, 21)
y_axis = binom.cdf(x_axis, 20, 0.12)
sns.barplot(x=x_axis, y=y_axis)
plt.show()
```
